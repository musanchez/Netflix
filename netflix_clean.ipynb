```python
import pandas as pd

#Step 1: Look into data

# 1. Print the first rows of your dataset
df = pd.read_csv('netflix_titles.csv')
display(df.head(7))

```

```python
# 2. Save the variables to a list
variables = df.columns.to_list()
display(pd.DataFrame(variables, columns=['Variables']))
```

```python
# 3. Note down potential issues you will have to address in each column
display(df.head(7))

# Print columns with missing values
# cast, director, countrys, etc have missing values
# date added is not in datetime format
# Description of the movie or series is not in the right format
# Duration is not in the right format
# There are columns that contain contatenated values
```

```python
#Step 2: Look at the proportion of missing data.

missing_columns = df.columns[df.isnull().any()]
missing_data = df[missing_columns].isnull().sum()
total_rows = df.shape[0]
missing_data = missing_data/total_rows*100
display(pd.DataFrame(missing_data, columns=['Missing Data (%)']))
```

```python
#Step 3: Look at the data types of each column
display(df.info()) #We can easily check the amount of missing values in each column
```

```python
#Step 4: If you have columns of Strings, check for trailing whitespaces
display(variables)
variables.remove('release_year')
df[variables] = df[variables].apply(lambda x: x.str.strip())
#This will remove trailing whitespaces from all columns
variables = df.columns.to_list()
```

```python
display(pd.DataFrame(variables, columns=['Variables']))
#Step 5: Dealing with missing values
missing_columns_string = ['director', 'cast', 'country', 'rating']
df[missing_columns_string] = df[missing_columns_string].fillna('Unknown')

#Current date format: month day, year
df['date_added'] = pd.to_datetime(df['date_added'], format='%B %d, %Y')
df['month_added'] = df['date_added'].dt.month
df['day_added'] = df['date_added'].dt.day
df['year_added'] = df['date_added'].dt.year
display(df.head(7))
```

```python
#Separate duration in two columns
df['duration_number'] = df['duration'].str.extract('(\\d+)')
df['duration_unit'] = df['duration'].str.extract('(\\D+)')
df['duration_unit'] = df['duration_unit'].str.strip()

# Convert duration_number column to numeric type
df['duration_number'] = pd.to_numeric(df['duration_number'])

# Replace 'Seasons' with 'Season' if duration_number is 1
df['duration_unit'] = df['duration_unit'].astype(str)
df.loc[df['duration_unit'] == 'Seasons', 'duration_unit'] = 'Season'
# Print the updated dataframe
display(df[['duration_number', 'duration', 'duration_unit']].head(7))
```

```python
#Step 7, check unique values

mask = df['rating'].str.contains(r'\d+\s*min', na=False)
df.loc[mask, 'duration'] = df.loc[mask, 'rating']
df.loc[mask, 'rating'] = 'Unknown'

df['rating'] = df['rating'].str.replace('Unknown', 'NR')
df['rating'] = df['rating'].str.replace('UR', 'NR')
display(pd.DataFrame(df['rating'].unique(),columns=['Unique Values']))
```

```python
#Unique countries and solution
df['country'] = df['country'].str.replace('West Germany', 'Germany')
df['country'] = df['country'].str.replace('Soviet Union', 'Russia')
df['country'] = df['country'].str.replace('East Germany', 'Germany')
unique_countries = set()
country_list = df['country'].unique().tolist()
for country in country_list:
    countries = country.split(',')
    for c in countries:
        c = c.strip()
        unique_countries.add(c)

unique_countries.remove('')
countries_df = pd.DataFrame(unique_countries, columns=['Unique Countries'])
countries_df = countries_df.sort_values(by='Unique Countries')

display(countries_df)
```

```python
display(df.loc[df['type'] == 'TV Show','type'].count())
display(df.loc[df['type'] == 'Movie','type'].count())
display(df[['listed_in', 'type']].head(7))

display(df.loc[df['listed_in'] == 'TV Shows','listed_in'].count())
display(df.loc[df['listed_in'] == 'Movies','listed_in'].count())
```

```python
#Step 8: Save the cleaned data
df.to_csv('cleaned_netflix_titles.csv', index=False)
```
